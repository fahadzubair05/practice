```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(10,6))

# plt.plot(gas.Year,gas.USA,'r.-', label = 'USA')
# plt.plot(gas.Year,gas.Canada,'b.-',label = 'Canada')
# plt.plot(gas.Year,gas.Australia,'g.-',label = 'Australia')
# plt.plot(gas.Year,gas.Germany,'y.-', label = 'Germany')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker = '.' , label = country)
    

plt.title('GAS PRICES COMPARISON')
plt.xticks(gas.Year[::3])

plt.xlabel('Years')
plt.ylabel('gallons/$')

plt.legend(bbox_to_anchor=(0.5, 1.0),
          ncol=3, fancybox=True, shadow=True)
plt.savefig('gas_prices_graph.png', dpi = 300)
plt.show()
```
